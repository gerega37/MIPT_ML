```python
import pandas as pd
```

```python
dataHERG1 = pd.read_csv('https://raw.githubusercontent.com/gerega37/mipt_study/main/dataHERG.csv', sep=';')
dataHERG1.head()
```

```python
dataHERG2 = pd.read_csv('https://raw.githubusercontent.com/gerega37/mipt_study/main/DOWNLOAD-beYKwKxjZ7zLbBnw2_3J6C0Ar_Tmp1Q6yJetu-ec8VA%3D.csv', sep=';')
dataHERG2
```

```python
result = pd.merge(dataHERG1, dataHERG2, on='Smiles')
```

```python
result.info()
```

```python
thresh = result.shape[0]*0.7
result = result.dropna(how='any', thresh=thresh, axis=1)
```

```python
dupl_columns = list(result.columns)

mask = result.duplicated(subset=dupl_columns)
result_duplicates = result[mask]
print(f'Число найденных дубликатов: {result_duplicates.shape[0]}')
```

```python
result = result.drop_duplicates(subset=dupl_columns)
print(f'Результирующее число записей: {result.shape[0]}')
```

```python
low_information_cols = []

for col in result.columns:
    top_freq = result[col].value_counts(normalize=True).max()
    nunique_ratio = result[col].nunique() / result[col].count()
    if top_freq > 0.95:
        low_information_cols.append(col)
        print(f'{col}: {round(top_freq*100, 2)}% одинаковых значений')
    if nunique_ratio > 0.95:
        low_information_cols.append(col)
        print(f'{col}: {round(nunique_ratio*100, 2)}% уникальных значений')
```

```python
result = result.drop(low_information_cols, axis=1)
print(f'Результирующее число признаков: {result.shape[1]}')
```

```python
result.info()
```

```python
result = result.drop(columns=[
    'AlogP_y',
    'Assay Organism',
    'Assay Cell Type',
    'Assay Description',
    'Source Description',
    'Assay Subcellular Fraction',
    'Molecular Weight_y',
    '#RO5 Violations_y',
    'Compound Key',
    'Standard Relation',
    'Inchi Key',
    'Structure Type']
)
```
